# 1. One-sample t-test

The mass of a sample of N = 20 acorns from a forest subjected to acid rain from a coal power plant are m = [8.8, 6.6, 9.5, 11.2, 10.2, 7.4, 8.0, 9.6, 9.9, 9.0, 7.6, 7.4, 10.4, 11.1, 8.5, 10.0, 11.6, 10.7, 10.3, and 7.0 g. ]

Does this sample provide enough evidence (alpha = 0.05) to say that the average mass of all acorns is different from 10 g?

**a) Formulate the null and alternate hypothesis**

Let  ùúá  be the average mass of all acorns.

We will test the null hypothesis

ùêª0:ùúá = 10

against the alternate hypothesis

ùêªùëé:ùúá ‚â† 10

**b) Calculate the test-statistic and based on the p-value provide a conclusion.**

```python
# define the array of the 20 samples acorns
m = np.array([8.8, 6.6, 9.5, 11.2, 10.2, 7.4, 8.0, 9.6, 9.9, 9.0, 7.6, 7.4, 10.4, 11.1, 8.5, 10.0, 11.6, 10.7, 10.3, 7.0 ])
print (m)
m.mean()
```

```python
#import the required functions
from scipy.stats import ttest_1samp

# calculate the test statistic and p-value
test_stat, p_value = ttest_1samp(m, popmean = 10, alternative = 'two-sided')
print('The p-value is ', p_value)
```

The p-value is less than the level of significance (0.05) and so there is enough statistical evidence to reject the null hypothesis. In other words, the average mass of all acorns is most probably different from 10, considering a level of significance of 0.05.


# 2. Independent (unpaired) two-sample t-test

The mass of N<sub>1</sub> = 20 acorns from oak trees up wind from a coal power plant and N<sub>2</sub> = 30 acorns from oak trees down wind from the same coal power plant are measured. Is the mass of acorns from trees down wind different from the¬†ones from up wind at a significance level of 0.05? The sample sizes are not equal but we will assume that the population variance for sample 1 and sample 2 are equal.

#### sample up wind:
x1 = [10.8, 10.0, 8.2, 9.9, 11.6, 10.1, 11.3, 10.3, 10.7, 9.7,
¬† ¬† ¬† 7.8, 9.6, 9.7, 11.6, 10.3, 9.8, 12.3, 11.0, 10.4, 10.4]

#### sample down wind:
x2 = [7.8, 7.5, 9.5, 11.7, 8.1, 8.8, 8.8, 7.7, 9.7, 7.0,
¬† ¬† ¬† 9.0, 9.7, 11.3, 8.7, 8.8, 10.9, 10.3, 9.6, 8.4, 6.6,
¬† ¬† ¬† 7.2, 7.6, 11.5, 6.6, 8.6, 10.5, 8.4, 8.5, 10.2, 9.2]

**a) Formulate the null and alternate hypothesis.**

Let  ùúá1,ùúá2  be the mean mass of up wind and down wind respectively.

We will test the null hypothesis

ùêª0:ùúá1 = ùúá2

against the alternate hypothesis

ùêªùëé:ùúá1 ‚â† ùúá2

**b) Calculate the test-statistic and based on the p-value provide a conclusion.**

```python
# define the arrays of the 20 samples acorns
x1 = np.array([10.8, 10.0, 8.2, 9.9, 11.6, 10.1, 11.3, 10.3, 10.7, 9.7, 7.8, 9.6, 9.7, 11.6, 10.3, 9.8, 12.3, 11.0, 10.4, 10.4])
x2 = np.array([7.8, 7.5, 9.5, 11.7, 8.1, 8.8, 8.8, 7.7, 9.7, 7.0, 9.0, 9.7, 11.3, 8.7, 8.8, 10.9, 10.3, 9.6, 8.4, 6.6, 7.2, 7.6, 11.5, 6.6, 8.6, 10.5, 8.4, 8.5, 10.2, 9.2])
print (x1)
print (x2)
```

```python
#import the required functions
from scipy.stats import ttest_ind

# find the p-value
test_stat, p_value = ttest_ind(x1, x2, equal_var = True, alternative = 'two-sided')
print('The p-value is ' + str(p_value))
```

The p-value is less than the level of significance. So we have enough statistical evidence to reject the null hypothesis. As such, we can conclude that the mass of acorns from trees up wind is different from that of acorns from trees down wind at a 0.05 level of significance.

# 3. ANOVA test

The marks obtained by 5 randomly picked students in Mathematics exam from three sections A, B, and C are as follows:

### Marks of 5 randomly picked students from Section A
A = [51, 45, 33, 45, 67]

### Marks of 5 randomly picked students from Section B
B = [23, 43, 23, 43, 45]

### Marks of 5 randomly picked students from Section C
C = [56, 76, 74, 87, 56]

Does the sample provide enough evidence to say that the mean marks of students in the three sections are different?

**a) Formulate the null and alternate hypothesis.**

Let  ùúá1,ùúá2,ùúá3  be the means of A, B, and C respectively.

We will test the null hypothesis

ùêª0:ùúá1=ùúá2=ùúá3

against the alternative hypothesis

ùêªùëé:  At least one of the mean scores is different from the rest.

**b) Calculate the test-statistic and based on the p-value provide a conclusion.**

```python
# define the arrays, each of students selected from one of the three sections
A = np.array([51, 45, 33, 45, 67])
B = np.array([23, 43, 23, 43, 45])
C = np.array([56, 76, 74, 87, 56])
```

Now, the normality and equality of variance assumptions need to be checked.

For testing of normality, Shapiro-Wilk‚Äôs test is applied to the response variable.

For equality of variance, Levene test is applied to the response variable.

### Shapiro-Wilk‚Äôs test
We will test the null hypothesis

ùêª0: Student marks follow a normal distribution

against the alternative hypothesis

ùêªùëé: Student markds do not follow a normal distribution

The  [shapiro()](https://colab.research.google.com/corgiredirector?site=https%3A%2F%2Fdocs.scipy.org%2Fdoc%2Fscipy%2Freference%2Fgenerated%2Fscipy.stats.shapiro.html) function of Scipy will be used to compute the test statistic and p-value.

```python
# Assumption 1: Normality
# import the required function
from scipy import stats

# find the p-value
w, p_value = stats.shapiro(np.array([A, B, C]))
print('The p-value is', p_value)
```

The p-value is large enough to reject the null hypothesis that the student marks follow a normal distribution.

### Levene‚Äôs test
We will test the null hypothesis

ùêª0 : All the population variances are equal

against the alternative hypothesis

ùêªùëé : At least one variance is different from the rest

The  [levene()](https://colab.research.google.com/corgiredirector?site=https%3A%2F%2Fdocs.scipy.org%2Fdoc%2Fscipy%2Freference%2Fgenerated%2Fscipy.stats.shapiro.html) function of Scipy will be used to compute the test statistic and p-value.

```python
#Assumption 2: Homogeneity of Variance
#import the required function
from scipy.stats import levene
statistic, p_value = levene( A, B, C)
# find the p-value
print('The p-value is', p_value)
```

Since the p-value is very large, we fail to reject the null hypothesis that the population variances are equal.

### Let's test whether the assumptions are satisfied or not

* The populations are normally distributed - Yes, the normality assumption is verified using the Shapiro-Wilk‚Äôs test.
* Samples are independent simple random samples - Yes, we are informed that the students were chosen randomly
* Population variances are equal - Yes, the homogeneity of variance assumption is verified using the Levene's test.

### Let's find the p-value

The [f_oneway()](https://colab.research.google.com/corgiredirector?site=https%3A%2F%2Fdocs.scipy.org%2Fdoc%2Fscipy%2Freference%2Fgenerated%2Fscipy.stats.f_oneway.html) function of Scipy will be used to compute the test statistic and p-value.

```python
#import the required function
from scipy.stats import f_oneway

# perform one-way anova test
test_stat, p_value = f_oneway(A, B, C)
print('The p-value is ' + str(p_value))
```

The p-value is less than the level of signficance (0.05) so we fail to reject the null hypothesis that all the three means are equal.

However, we don't know which mean is different from the rest or whether all pairs of means are different. Multiple comparison tests are used to test the differences between all pairs of means.

### Multiple Comparison test (Tukey HSD)

In order to identify for which fuel type mean carbon emission is different from other groups, the null hypothesis is

ùêª0:ùúá1=ùúá2 and ùúá1=ùúá3 and ùúá2=ùúá3

against the alternative hypothesis

ùêªùëé:ùúá1‚â†ùúá2 or ùúá1‚â†ùúá3 or ùúá2‚â†ùúá3

The  [pairwise_tukeyhsd()](https://colab.research.google.com/corgiredirector?site=https%3A%2F%2Fwww.statsmodels.org%2Fdev%2Fgenerated%2Fstatsmodels.stats.multicomp.pairwise_tukeyhsd.html)function of Statsmodels will be used to compute the test statistic and p-value.

```python
#import the required function
from statsmodels.stats.multicomp import pairwise_tukeyhsd

#create a dictionary of the arrays/lists
data = {'A': [51, 45, 33, 45, 67],
        'B': [23, 43, 23, 43, 45],
        'C': [56, 76, 74, 87, 56]}

#create empty row
rows = []

#Iterate over the dictionary items and arrays
for key, values in data.items():
    for value in values:
        rows.append([key, value])

#Convert the rows list to a DataFrame
df = pd.DataFrame(rows, columns=['Keys', 'Values'])

# perform multiple pairwise comparison (Tukey HSD)
m_comp = pairwise_tukeyhsd(endog = df['Values'], groups = df['Keys'], alpha = 0.05)
print(m_comp)
```

Since the p-value (p-adj) for the the section pairs A-C and B-C is lower than the level of significance (0.05), we can conclude that the mean marks for students are appreciably different for these section pairs.
