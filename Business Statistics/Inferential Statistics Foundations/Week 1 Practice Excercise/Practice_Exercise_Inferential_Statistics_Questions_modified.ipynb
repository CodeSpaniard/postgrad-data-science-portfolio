# Binomial Distribution

### Q1. A LED bulb manufacturing company regularly conducts quality checks at specified periods on the products it manufactures.  Historically, the failure rate for LED light bulbs that the company manufactures is 5%. Suppose a random sample of 10 LED light bulbs is selected. Find the probability distribution for the defective bulbs and answer the following questions.

**a) What is the probability that none of the LED bulbs are defective?**

```python
n = 10 # the sample size
p = 0.05 # declare p which represents the probability of the LED bulb being defective
k = np.arange(0,11) # different possible number of defective bulbs
from scipy.stats import binom # import the required function

# use the binom.pmf() function to generate the probability distribution
binomial = binom.pmf(k=k, n=n, p=p)

# plot the distribution
plt.bar(k, binomial) # make a bar plot
plt.title("Binomial: n=%i , p=%.2f" % (n, p), fontsize=15) # set the title
plt.xlabel("Number of Defective bulbs") # set the x-axis label
plt.ylabel("Probability of bulbs being defective") # set the y-axis label
plt.show() # display the plot
```

```python
binomial[0] # probability that none of the bulbs is defective
```

**b) What is the probability that exactly one of the LED bulbs is defective?**

```python
binomial[1] # probability that exactly one of the bulbs is defective
```

**c) What is the probability that two or fewer of the LED bulbs are defective?**

Hint: We need to calculate cumulative probability of two or fewer LED bulbs being defective.

```python
binom.cdf(k=2, n=n, p=p) # probability that two or fewer of the bulbs are defective
```

**d) What is the probability that three or more of the LED bulbs are defective?**

Hint: We need to subtract the cumulative probability of up to 2 defective LED bulbs from 1.

```python
1 - binom.cdf(k=2, n=n, p=p) # probability that three or more of the bulbs are defective
```

### Q3.	Over a long period in a large multinomial corporation, 10% of all sales trainees are rated as outstanding, 75% are rated as excellent, 10% are rated as satisfactory and 5% are considered unsatisfactory. Find the following probabilities for a sample of 10 trainees selected at random.

**a) Two are rated as outstanding**

```python
binom.pmf(k=2, n=10, p=0.10) # probability that two of the trainees are rated outstanding
```

19.37% probability that two trainees are rated as outstanding

**b) Two or more are rated as outstanding**

```python
1 - binom.cdf(k=1, n=10, p=0.10) # probability that two or more trainees are rated as outstanding
```

26.39% probability that two or more trainees are rated as outstanding

**c) Eight of the ten are rated either outstanding or excellent**

```python
p_out_or_exc = 0.10 + 0.75 # the probability of being rated outstanding or excellent
                           # is the sum of the individual probabilities since they are mutually exclusive
binom.pmf(k=8, n=10, p=p_out_or_exc) # probability that eight of the ten are rated either outstanding or excellent
```

27.59% probability that eight of the ten are rated either outstanding or excellent

**d) None of the trainees are rated as unsatisfactory**

```python
binom.pmf(k=0, n=10, p=0.05) # probability that none of the trainees are rated outstanding
```

59.87% probability that none of the trainees are rated as unsatisfactory

# Normal Distribution

### Q4. According to the Telecommunication Industry, the average monthly cell phone bill is Rs. 850 with a standard deviation of Rs. 150. Assuming that the monthly bill follows a normal distribution, answer the following questions.

```python

mu = 850 # mean
sigma = 150 # standard deviation
print("The mean is", mu)
print("The standard deviation is", sigma)
```

**a) What is the probability that a randomly selected cell phone bill is more than Rs 1200?**

```python
# import the required function
from scipy.stats import norm
prob_more_than_1200 = 1 - norm.cdf(1200, mu, sigma) # probability that a randomly selected cell phone bill is more than Rs 1200
prob_more_than_1200
```

0.98% probability that a randomly selected cell phone bill is more than Rs 1200

**b) What is the probability that a randomly selected cell phone bill is between Rs 750 and Rs 1200?**

```python
norm.cdf(1200, mu, sigma) - norm.cdf(750, mu, sigma) # probability that a randomly selected cell phone bill is between Rs 750 and Rs 1200
```

73.77% probability that a randomly selected cell phone bill is between Rs 750 and Rs 1200

**c) What is the probability that a randomly selected cell phone bill is no more than Rs 650?**

```python
norm.cdf(650, mu, sigma) # probability that a randomly selected cell phone bill is less than or equal to Rs 650
```

9.12% probability that a randomly selected cell phone bill is less than or equal to Rs 650

**d) What is the amount above which lies top 15% of cell phone bills?**

```python
round(norm.ppf(0.85, mu, sigma)) #85th percentile
```

The 85th percentile is 1005

**e) What is the amount below which lies bottom 25% of cell phone bills?**

```python
round(norm.ppf(0.25, mu, sigma)) #25th percentile
```

The 25th percentile is 749

### Q5. The mean filling capacity for a coke bottle is 500 ml with a standard deviation of 20 ml. Assume that it follows a normal distribution and answer the following questions.

**a) What is the probability that the bottle filled less than 480 ml?**

```python
mu = 500 # mean
sigma = 20 # standard deviation
norm.cdf(480, mu, sigma) # probability that the bottle filled less than 480 ml
```

15.87% probability that the bottle filled less than 480 ml

**b) What is the probability that the bottle filled more than 520 ml?**

```python
1 - norm.cdf(520, mu, sigma) # probability that the bottle filled more than 520 ml
```

15.87% probability that the bottle filled more that 520 ml

**c) What is the probability that the bottle filled between 470 ml to 525 ml?**

```python
norm.cdf(525, mu, sigma) - norm.cdf(470, mu, sigma) # probability that the bottle filled between 470 ml to 525 ml
```

82.75% probability that the bottle filled between 470 ml to 525 ml

### Q6. In 2 Liter soft drink bottles, the drink filled is normally distributed, with a mean of 2.0 liters and a standard deviation of 0.05 liter. If bottles contain less than 95% of the listed net content (1.90 liters), the manufacturer may be subject to penalty by the state office of consumer affairs. Bottles that have a net quantity above 2.1 liters may cause excess spillage upon opening. Answer the following questions.

**a) What is the probability that the bottle content is between 1.9 and 2.0 liters?**

```python
mu = 2.0 # mean
sigma = 0.05 # standard deviation
norm.cdf(2.0, mu, sigma) - norm.cdf(1.9, mu, sigma) # probability that the bottle content is between 1.9 and 2.0 liters
```

47.72% probability that the bottle content is between 1.9 and 2.0 liters

**b) What is the probability that the bottle content is between 1.9 and 2.1 liters?**

```python
norm.cdf(2.1, mu, sigma) - norm.cdf(1.9, mu, sigma) # probability that the bottle content is between 1.9 and 2.1 liters
```

95.45% probability that the bottle content is between 1.9 and 2.1 liters

**c) What is the probability that the bottle content is below 1.9 liters or above 2.1 liters?**

```python
norm.cdf(1.9, mu, sigma) + 1 - norm.cdf(2.1, mu, sigma) # probability that the bottle content is below 1.9 or above 2.1 liters
```

4.55% probability that the bottle content is below 1.9 or above 2.1 liters

**d) 99% of the bottles contains at least what amount of soft drink in them?**

```python
round(norm.ppf(0.99, mu, sigma)) #99th percentile
```

The 99th percentile is 2 liters
