```python
# Libraries to help with reading and manipulating data
import numpy as np
import pandas as pd

# Libraries to help with data visualization
import matplotlib.pyplot as plt
import seaborn as sns

# to restrict the float value to 3 decimal places
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
data = pd.read_csv('/content/drive/MyDrive/Python Course/Exploratory Data Analysis/Titanic+Data+Set.csv')
```

```python
data.shape
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data = data.drop(['Cabin'],axis=1) #or df.drop('Cabin', axis = 1, inplace = True)
data.isnull().sum()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=data,x='Age',y='Sex', hue='Survived')
plt.show()
```

```python
sns.histplot(data=data,x='Parch',stat='density')
plt.show()
sns.boxplot(data=data,x='Parch')
plt.show()
```

```python
sns.countplot(data=data,x='Pclass', hue='Survived')
plt.xticks(rotation=90)
plt.show()
```

```python
data.groupby(['Pclass'])['Survived'].sum()
```

```python
sns.countplot(data=data,x='Embarked', hue='Survived')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.catplot(x='Fare',
            col='Sex',
            data=data,
            col_wrap=4,
            kind="box")
plt.show()
```
