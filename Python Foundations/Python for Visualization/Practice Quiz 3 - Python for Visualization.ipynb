```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


np.random.seed(42)
data1 = np.random.normal(loc=5, scale=1, size=300)
data2 = np.random.normal(loc=10, scale=2, size=700)
data = np.concatenate([data1, data2])

# True density function
x_values = np.linspace(0, 20, 1000)
true_density = 0.3 * norm.pdf(x_values, loc=5, scale=1) + 0.7 * norm.pdf(x_values, loc=10, scale=2)

plt.plot(x_values, true_density, label='True Density')
plt.hist(data, bins=20, density=True, alpha=0.5, label='Histogram')
plt.legend()
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('True Density and Histogram')
plt.show()
```

```python
# Gaussian kernel function
def gaussian_kernel(x, x_i, bandwidth):
    return np.exp(-(x - x_i)**2 / (2 * bandwidth**2)) / (np.sqrt(2 * np.pi) * bandwidth)

bandwidth = 1.0
bin_width = 1
bins = np.arange(0, 20, bin_width)
kde_density = np.zeros_like(bins, dtype=float)

for i, x in enumerate(bins):
    for x_i in data:
        kde_density[i] += gaussian_kernel(x, x_i, bandwidth)
    kde_density[i] /= (len(data) * bandwidth * bin_width)


plt.plot(x_values, true_density, label='True Density')
plt.hist(data, bins=20, density=True, alpha=0.5, label='Histogram')
plt.plot(bins, kde_density, label='KDE Estimate', linestyle='dashed')
plt.legend()
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('True Density, Histogram, and KDE Estimate')
plt.show()
```
